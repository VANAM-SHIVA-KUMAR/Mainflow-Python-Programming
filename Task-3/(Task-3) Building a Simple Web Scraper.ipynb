```python
import requests
from bs4 import BeautifulSoup

# URL of the page you want to scrape
url = 'https://en.wikipedia.org/wiki/Web_scraping'  # Example URL for web scraping

# Headers to mimic a real browser
headers = {'User-Agent': 'Mozilla/5.0'}

# Fetch the web page
try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Check if the request was successful
    page_content = response.content
except requests.exceptions.RequestException as e:
    print(f"Error fetching the page: {e}")
    page_content = None

if page_content:
    # Parse the HTML content
    soup = BeautifulSoup(page_content, 'html.parser')

    # Extract text
    text = soup.get_text()
    print("Text from the page:")
    print(text)

    # Extract links
    links = [a['href'] for a in soup.find_all('a', href=True)]
    print("\nLinks from the page:")
    for link in links:
        print(link)

    # Extract images
    images = [img['src'] for img in soup.find_all('img', src=True)]
    print("\nImages from the page:")
    for image in images:
        print(image)

    # Output links to a file
    with open('links_output.txt', 'w') as f:
        for link in links:
            f.write(f"{link}\n")

    print("\nLinks have been written to links_output.txt")

# Reference for BeautifulSoup
print("\nFor more information on BeautifulSoup, visit: https://www.crummy.com/software/BeautifulSoup/bs4/doc/")
```
