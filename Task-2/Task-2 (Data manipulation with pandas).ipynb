```python
import pandas as pd

# Read CSV file
df = pd.read_csv(r"D:\Main Flow Internship\example.csv")

# View the first few rows
print(df.head())

# Get basic statistics
print(df.describe())

# Select a column
age_column = df['Age']
print(age_column)

# Filter rows where Salary is greater than 60000
high_salary_df = df[df['Salary'] > 60000]
print(high_salary_df)

# Add a new column for Bonus (10% of Salary)
df['Bonus'] = df['Salary'] * 0.1

# Drop the Gender column
df = df.drop('Gender', axis=1)

# Handle missing data (if any)
df = df.fillna(0)

# Save to a new CSV file
df.to_csv('modified_example.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# Read CSV file
df = pd.read_csv(r"D:\Main Flow Internship\example.csv")

# Introduce missing values (for demonstration)
df.loc[2, 'Age'] = np.nan
df.loc[4, 'Salary'] = np.nan

# Introduce duplicate rows (for demonstration)
df = df.append(df.iloc[0], ignore_index=True)
df = df.append(df.iloc[2], ignore_index=True)

# Check for missing values
print("Missing values before cleaning:")
print(df.isnull().sum())

# Fill missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# Check for missing values after cleaning
print("Missing values after cleaning:")
print(df.isnull().sum())

# Check for duplicate rows
print("Duplicates before removing:")
print(df.duplicated().sum())

# Remove duplicate rows
df_cleaned = df.drop_duplicates()

# Check for duplicate rows after cleaning
print("Duplicates after removing:")
print(df_cleaned.duplicated().sum())

# Display the cleaned DataFrame
print(df_cleaned)

# Save the cleaned DataFrame to a new CSV file
df_cleaned.to_csv('cleaned_example.csv', index=False)
```

```python
import pandas as pd

# Read CSV file
df = pd.read_csv(r"D:\Main Flow Internship\example.csv")

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Filtering Data
filtered_df = df[df['Age'] > 30]
print("\nFiltered DataFrame (Age > 30):")
print(filtered_df)

# Sorting Data
sorted_df = df.sort_values(by='Salary', ascending=False)
print("\nSorted DataFrame (by Salary, descending):")
print(sorted_df)

# Grouping Data
grouped_df = df.groupby('Gender')['Salary'].mean().reset_index()
print("\nGrouped DataFrame (mean Salary by Gender):")
print(grouped_df)
```
